```python
import pandas as pd
from pandas import json_normalize
import pickle
```

```python
# Load players information
with open("ch_summoner","rb") as f:
    ch_summoner = pickle.load(f)

with open("gm_summoner","rb") as f:
    gm_summoner = pickle.load(f)
```

```python
ch_summonerId = ch_summoner['summonerId']
gm_summonerId = gm_summoner['summonerId']
ch_summonerId
gm_summonerId
```

```python
with open("ch_puuid","rb") as f:
    ch_puuid = pickle.load(f)

with open("gm_puuid","rb") as f:
    gm_puuid = pickle.load(f)
```

```python
ch_puuid
gm_puuid
```

```python
Ch_summoner_table = pd.DataFrame(ch_summoner)
GM_summoner_table = pd.DataFrame(gm_summoner)
```

```python
Ch_summoner_table["puuid"] = list(ch_puuid)
Ch_summoner_table["rank"] =  ["Challenger" for i in range(len(ch_puuid))]
Ch_summoner_table["win_rate (%)"] =  [Ch_summoner_table.iloc[i]["wins"]/(Ch_summoner_table.iloc[i]["wins"]+Ch_summoner_table.iloc[i]["losses"])*100 for i in range(len(ch_puuid))]
Ch_summoner_table = Ch_summoner_table[["puuid", "summonerId", "summonerName", "rank", "leaguePoints", "wins", "losses", "win_rate (%)" ]]
Ch_summoner_table = Ch_summoner_table.sort_values("leaguePoints", ascending=False)
Ch_summoner_table.head(60)
```

```python
pd.set_option('display.max_rows', None) 
Ch_summoner_table
```

```python
GM_summoner_table["puuid"] = list(gm_puuid)
GM_summoner_table["rank"] =  ["GrandMaster" for i in range(len(gm_puuid))]
GM_summoner_table["win_rate (%)"] =  [GM_summoner_table.iloc[i]["wins"]/(GM_summoner_table.iloc[i]["wins"]+GM_summoner_table.iloc[i]["losses"])*100 for i in range(len(gm_puuid))]
GM_summoner_table = GM_summoner_table[["puuid", "summonerId", "summonerName", "rank", "leaguePoints", "wins", "losses", "win_rate (%)" ]]
GM_summoner_table = GM_summoner_table.sort_values("leaguePoints", ascending=False)
GM_summoner_table
```

```python
summoner_table = pd.concat([Ch_summoner_table,GM_summoner_table], ignore_index=True)
summoner_table
```

```python
summoner_table.to_csv("summoner_table.csv")
```

```python
game = pd.read_csv('match_record_050422023.csv', index_col=[0])
game
```

```python
import re
pattern = re.compile(r'Version 13\.6\.499\.7758')
gmae_table = game[game['info.game_version'].str.contains(pattern)]
gmae_table
```

```python
from datetime import datetime

game_data = gmae_table[["info.tft_game_type"]]

game_id_list = []
game_part_list = []
game_date_list = []
game_time_list = []

for i in range(gmae_table.shape[0]):
    game_id_list.append(gmae_table.iloc[i]["metadata.match_id"])
    game_part_list.append(gmae_table.iloc[i]["metadata.participants"])
    game_date_list.append(str(datetime.fromtimestamp(int(gmae_table.iloc[i]['info.game_datetime'])/1000)).split(' ')[0])
    game_time_list.append(int(gmae_table.iloc[i]["info.game_length"])/60)

game_data["game_id"] = game_id_list
game_data["metadata.participants"] = game_part_list
game_data["game_date"] = game_date_list
game_data["ingame_time (m)"] = game_time_list

game_data
```

```python
game_data = game_data[["game_id", "metadata.participants", "game_date", "ingame_time (m)", "info.tft_game_type"]]
game_data
```

```python
Not_standard = game_data['info.tft_game_type'] != 'standard'
Not_standard
```

```python
game_data.to_csv("game_data.csv")
```
